## Correction step 2/2  - predictionstring 교정 - 진행중

```python
import re
import string
import os

import pandas as pd
import numpy as np
from tqdm import tqdm
from pathlib import Path

BASE_DIR = Path('../input/feedback-prize-2021')
TRAIN_DIR = BASE_DIR / 'train'

```

```python
tr_df = pd.read_csv(BASE_DIR / 'correct_v1.csv')
# df = pd.read_csv(BASE_DIR + "train.csv")

tr_df.head(5)
```

```python
"""
correction #2 : cmp with discTxtf in train.csv and predStr word from train dir
"""

def get_tokens(ids):
    with open(TRAIN_DIR/f'{ids}.txt', 'r', encoding='UTF-8') as file: data = file.read()
    return data.split()

def get_text(ids):
    with open(TRAIN_DIR/f'{ids}.txt', 'r', encoding='UTF-8') as file: data = file.read()
    return data

def verify_text(row):
    tokens = get_tokens(row['id'])
    indices = [int(x) for x in row.predictionstring.split()]
    start = indices[0]
    end = indices[-1] + 1
    text = ' '.join(tokens[start:end])
    return text

verify_text(tr_df.loc[120722])

```

```python
# train_names, train_texts = [], []
# for f in tqdm(list(os.listdir(TRAIN_DIR))):
#     train_names.append(f.replace('.txt', ''))
#     train_texts.append(open(TRAIN_DIR/f, 'r', encoding='UTF-8').read())

# train_texts = pd.DataFrame({'id': train_names, 'text': train_texts})
# train_texts

tqdm.pandas()
tr_df['verify'] = tr_df.progress_apply(verify_text, axis=1)
```

```python
tr_df['correct'] = tr_df.progress_apply(lambda row: ' '.join(row.discourse_text.split()) == row.verify, axis=1)
tr_df.correct.value_counts()
```

```python
for _, row in tr_df[tr_df.correct == False].sample(n=5).iterrows():
    print(row['id'])
    print(row.discourse_text)
    print(row.verify)
```
